```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.models import load_model
from keras.utils import np_utils

image1 = cv2.imread("./test_img_01.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image1)
```

```python
ret, img_threshold = cv2.threshold(image1, 143,255,cv2.THRESH_BINARY)
resize = cv2.resize(img_threshold,(28,28))
plt.imshow(resize)
```

```python
test = np.resize(resize,(1,28,28))*(-1)


model = load_model('ImageDatagen_3.h5')

res = model.predict(test)
res = res.argmax()
print(res)
```
